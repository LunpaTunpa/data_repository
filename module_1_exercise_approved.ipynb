## 2\. Números

Preencha as células de código para preencher os valores de (A), (B) e (C) na tabela de ticket médio abaixo:

<br>

| Dia   | Valor Total Vendas | Qtd Total Vendas | Ticket Medio |
|-------|--------------------|------------------|--------------|
| 19/01 | (A)                | 3                | 320.52       |
| 20/01 | 834.47             | (B)              | 119.21       |
| 23/01 | 15378.12           | 5                | (C)          |

```
A = 320.52 * 3; # Pegue o valor do Ticket Médio e multiplique pela quantidade total de vendas para obter A;
print (f'{A:.2f}'); # Mostre a variavel A, limitando-a apenas a duas casas decimais através da função f;
```

```
B = 834.47 / 119.21; # Pegue o valor Total das vendas e dividao pelo valor do Ticket Médio p/ obter B;
print (f'{B:.0f}'); # Mostre a variavel B, limitando-a 0 casas decimais através da função f;
```

```
C = 15378.12 / 5; # Pegue o valor Total das vendas e dividao pela quantidade Total de Vendads p/ obter C;
print (f'{C:.2f}'); # Mostre a variavel C, limitando-a apenas a duas casas decimais através da função f;
```

## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'
mcancao = cancao.upper(); # Torne todos os caracteres de "cancao" máusculos e insira o resultado em "mcancao";
print(mcancao.count('RODA')); # Conte as vezes que o caracter "RODA" aparece em "mcancao";
print(mcancao.replace('RODA', 'GIRA')); # Troque o caracter "RODA" por "GIRA" em "mcancao" e mostre na tela;
```

Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

```
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'
print(noticia.find('2')); # Encontre a posição do caracter "2";
print(noticia.find('%')); # Encontre a posição do caracter "%";
print(noticia.find('6')); # Encontre a posição do caracter "6";

selic = noticia[12:17]; # Insira os caracteres no trecho do caracter 12 ao 17 dentro da variável "selic";
ano = noticia[62]; # Insira o caractere na posição "62" dentro da variável "ano";

print('A nova taxa Selic é de', selic); # Mostre ...
print('Já fazem', ano, 'anos que a taxa não vai as alturas.'); # Mostre ...
```

## 4\. Booleanos

Utilize a tabela da verdade para responder: qual o valor da variável x?

```
a = False
b = True

x = not a & b # Valor "inverso" de x;

print('Sendo inverso de:')
print(a, "e", b, '=', x);

x = a & b; # Valor "original" de x;

print('Sendo:')
print(a, "e", b, '=', x);
```
